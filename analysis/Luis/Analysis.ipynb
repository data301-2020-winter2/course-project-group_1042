```python
from project_functions import *
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = data_processing_rename("../../data/raw/Medical_Cost.csv")
df.head()
```

```python
# Creating dataframes with different sorts for easy loc functions
from project_functions import *
df_age = data_processing_age_sort("../../data/raw/Medical_Cost.csv")
df_region = data_processing_region_sort("../../data/raw/Medical_Cost.csv")
df_children = data_processing_children_sort("../../data/raw/Medical_Cost.csv")
df_smoker = data_processing_smoker_sort("../../data/raw/Medical_Cost.csv")
df_sex = data_processing_sex_sort("../../data/raw/Medical_Cost.csv")
```

```python
# Obtaining values for the original dataset
print('Original Dataset')
print('Count:', df['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (df['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (df['Medical Costs'].sum()))

# Obtaining values for smokers
temp = df_smoker.drop(df_smoker[df_smoker.Smoker == 'no'].index).reset_index(drop = True)
print('\nOnly Smokers')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for non-smokers
temp = df_smoker.drop(df_smoker[df_smoker.Smoker == 'yes'].index).reset_index(drop = True)
print('\nOnly Non Smokers')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for people with children
temp = df_children.drop(df_children[df_children.Children > 0].index).reset_index(drop = True)
print('\nOnly People With Children')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for people without children
temp = df_children.drop(df_children[df_children.Children == 0].index).reset_index(drop = True)
print('\nOnly People With No Children')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for men
temp = df_children.drop(df_sex[df_sex.Sex == 'female'].index).reset_index(drop = True)
print('\nOnly Men')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for women
temp = df_children.drop(df_sex[df_sex.Sex == 'male'].index).reset_index(drop = True)
print('\nOnly Women')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for people in the northeast
temp = df_region.drop(df_region[df_region.Region != 'northeast'].index).reset_index(drop = True)
print('\nOnly People In The Northeast')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for people in the northwest
temp = df_region.drop(df_region[df_region.Region != 'northwest'].index).reset_index(drop = True)
print('\nOnly People In The Northwest')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for people in the southeast
temp = df_region.drop(df_region[df_region.Region != 'southeast'].index).reset_index(drop = True)
print('\nOnly People In The Southeast')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))

# Obtaining values for people in the southwest
temp = df_region.drop(df_region[df_region.Region != 'southwest'].index).reset_index(drop = True)
print('\nOnly People In The Southwest')
print('Count:', temp['Medical Costs'].count(), 'entries')
print('Mean: %.2f dollars' % (temp['Medical Costs'].mean()))
print('Sum: %.2f dollars' % (temp['Medical Costs'].sum()))
```
