```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
df = pd.read_csv('../../data/raw/Medical_Cost.csv')
df.rename(columns = str.title, inplace = True)
```

```python
pd.DataFrame.count(df)
df.describe()
```

```python
#Sorting data
female = df.loc[df['Sex']=='female']
male = df.loc[df['Sex']=='male']
smoker = df.loc[df['Smoker']=='yes']
non_smoker = df.loc[df['Smoker']=='no']
bin_bmi = pd.qcut(df['Bmi'], 4)
```

```python
#Boxplot Comparison of charges between gender and smoking
sns.boxplot(x='Smoker', y='Charges', data=df, hue='Sex')
```

```python
#Violin Plot Comparison of charges between gender and smoking
sns.violinplot(x='Smoker', y='Charges', data=df, hue='Sex', split=True, type='paper')
```

```python
#Scatterplots relating BMI to medical costs
df_grid = sns.FacetGrid(df, col='Sex', row='Smoker')
df_grid.map(plt.scatter, 'Bmi', 'Charges', marker='.')
```
